```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import json
import re
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

stop = stopwords.words('english')
def panduan(i):
    if i>=0.05:
        s='Positive'
    elif i<= -0.05:
        s='Negative'
    else:
        s='Neutral'
    return s

def sentiment_scores(sentence_list):
	# Create a SentimentIntensityAnalyzer object.
	sid_obj = SentimentIntensityAnalyzer()
	sentiment=[]
	for sentence in sentence_list:
		sentence_dict = sid_obj.polarity_scores(sentence)
		sentiment.append(sentence_dict['compound'])
	return sentiment
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()
lgbt=['genderqueer','pansexual','asexual','intersex','two-spirit','closeted','out and proud','lgbt','lgbtq','pride','ally','queer','gay','lesbian','bisexual','transgender','non-binary','coming out','same-sex','homophobia','heteronormativity','rainbow','stonewall','out and proud','genderfluid','' ]
transport=['car','shipping','park','bicycle','subway','metro','train','tram','airport','bridge','road','lane','intersection','vehicle','transport','public transportation','commute','driver','passenger','traffic','highway','accident','crash','one-way','station','crowd','crossing','speed']
rent=['rental insurance','rental application','agent','lease','property','application','occupancy','inspection','rental market','utility','sublease','deposit','rent','tenant','landlord','accommodation','apartment','house','studio','1b1b','2b1b','2b2b']
lgbt = [lemmatizer.lemmatize(token) for token in lgbt]
lgbt = [stemmer.stem(token) for token in lgbt]
transport = [lemmatizer.lemmatize(token) for token in transport]
transport = [stemmer.stem(token) for token in transport]
rent = [lemmatizer.lemmatize(token) for token in rent]
rent = [stemmer.stem(token) for token in rent]
# lgbt=['lgbt','lgbtq','pride','ally','queer','gay','lesbian','bisexual','transgender','non-binary','coming out','same-sex','homophobia','heteronormativity','rainbow','stonewall' ]
def text_similarity(text1):
    s=text1.lower()
    # Delete the @
    s = re.sub(r"@[A-Za-z0-9]+", ' ', s)
    # Delete URL links
    s = re.sub(r"https?://[A-Za-z0-9./]+", ' ', s)
    # Just keep letters and important punctuation
    s = re.sub(r"[^a-zA-Z']", ' ', s)
    # Remove additional spaces
    s = re.sub(r" +", ' ', s)
    s=' '.join( [w for w in s.split() if len(w)>1] )
    s = re.sub(r'[^\w\s]', '', s)
    # Tokenize and lemmatize the texts
    tokens1 = word_tokenize(s)
    # tokens2 = word_tokenize(text2)
    lemmatizer = WordNetLemmatizer()
    stemmer = PorterStemmer()
    tokens1 = [lemmatizer.lemmatize(token) for token in tokens1]
    tokens1 = [stemmer.stem(token) for token in tokens1]
    # Remove stopwords
    stop_words = stopwords.words('english')
    tokens1 = [token for token in tokens1 if token not in stop_words]
    c = 0
    if any(keyword in tokens1 for keyword in lgbt):
        c=1
    elif any(keyword in tokens1 for keyword in transport):
        c=2
    elif any(keyword in tokens1 for keyword in rent):
        c=3
    
    return c

def clean_twitter(s):
    s=s.lower()
    # Delete the @
    s = re.sub(r"@[A-Za-z0-9]+", ' ', s)
    # Delete URL links
    s = re.sub(r"https?://[A-Za-z0-9./]+", ' ', s)
    # Just keep letters and important punctuation
    s = re.sub(r"[^a-zA-Z.!?']", ' ', s)
    # Remove additional spaces
    s = re.sub(r" +", ' ', s)
    s=" ".join([word for word in s.split() if word not in (stop)])
    stemmer = PorterStemmer()
    lemmatizer = WordNetLemmatizer()
    for word in s:
          stemmer.stem(word)
          lemmatizer.lemmatize(word) 
    cleaned_s=' '.join( [w for w in s.split() if len(w)>1] )
    return cleaned_s

def clean(s):
    a=s
    if '\"' in s:
        delimiter = '"'
        a=s.split(delimiter)[0]
    a=s.split(",")[0]
    return a
```
